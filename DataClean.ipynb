```python
def getcleandept(dept):
    removeterms = ['(Photomedia)','Interdisciplinary','Develop),','UW','Bothell','Departmental','Honors','College','Tacoma','(General)','(Journalism)','(Jazz Studies)']
    if dept.endswith(";"):
        dept = dept.rstrip(";")
    dept = dept.replace('&amp',"")
    dept = dept.replace('\xc2\xa0',"")
    dept = ' '.join([i for i in dept.split() if i not in removeterms]).strip()
    if ";" in dept:
        depts = dept.split(";")
    else:
        depts = [dept]
    finaldepts = []
    
    if depts == ['Aquatic ', ' Fishery Sciences']:
        depts = ['Aquatic and Fishery Science']
    elif depts == ['Biochemistry and Anthropology']:
        depts = ['Biochemistry','Anthropology']
    
    for tmpdept in depts:
        if tmpdept.endswith(","):
            tmpdept = tmpdept.rstrip(',')
        
        #remove trailing whitespaces
        tmpdept = tmpdept.strip()
        
        #dealing with spellings
        if tmpdept == 'Mathmatics' or tmpdept == 'Mathmatics' or tmpdept=="Math":
            tmpdept = "Mathematics"
        elif tmpdept == 'Applied Math and Computational Science':
            tmpdept = "Applied Mathematics"
        elif 'Music' in tmpdept:
            tmpdept = 'Music'
        
        #dealing with multiple mentions of dept
        if "Biology" in tmpdept:
            tmpdept = "Biology"
        elif tmpdept == 'Law, Societies, and Justice':
            tmpdept = 'Law, Society and Justice'
        
        
        if tmpdept !='':
            if tmpdept == 'Mathmatics':
                print depts
            finaldepts.append(tmpdept)
    return finaldepts
```

```python
finalcleandepts = []
for dept in df.majors.tolist():
    finalcleandepts.append(getcleandept(dept))
```

```python
uniquedepts = set()
for depts in finalcleandepts:
    for tmpdept in depts:
        uniquedepts.add(tmpdept)
```

```python
df['cleandepts'] = finalcleandepts
df.to_csv("CLEANDATA.csv")
```
